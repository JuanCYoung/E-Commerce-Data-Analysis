### Customer and Order dataframe combined

```python
customer_joined_df = pd.merge(
    left = customers_df, 
    right = orders_df, 
    how = "right",
    on = "customer_id"
)
```

```python
customer_joined_df[customer_joined_df.order_delivered_customer_date.isna()]
```

The presence of NA values in some columns does not necessitate removal or dropping. In the context of orders, it is a normal occurrence for certain columns to have NA values. For instance, in cases where an order is canceled, it is expected that the order_delivered_date becomes NA since no shipment takes place.

### Order_items, Order and Order_payments combined 

```python
order_items_pay_df = pd.merge(
    left=pd.merge(
        left=order_items_df,
        right=order_payments_df,
        how="inner",
        on="order_id"),
    right=orders_df,
    how="inner",
    on="order_id"
)
```

```python
order_items_pay_df[order_items_pay_df.order_delivered_customer_date.isna()]
```

Similar to the previous case, a null value is considered normal in the context of an order. When the status is "shipped," the order_delivered_customer_date is expected to be null because the product has not yet been delivered to the customer.
